```python
import os
import pandas as pd

path= r"C:\Users\Lenovo\Desktop\Datasets_for_ML_-_Regression"
os.chdir(path)
os.listdir(path)
data=pd.read_csv("TShirt_size.csv")
```

```python
data.head()
import numpy as np
```

```python
data1=np.sqrt((data['Height (in cms)']-161)**2 + (data['Weight (in kgs)']-61)**2)
```

```python
data['Euclidean_destance'] = data1
data.sort_values(by='Euclidean_destance',ascending=True)
```

```python
Value=int(input("Enter K Value"))
print("Prediction:",str(data['T Shirt Size'][:Value].mode())[5])
```

```python
d = {'Age':pd.Series([25,35,45,20,35,52,23,40,60,48,33]),
   'Loan Amount':pd.Series([4000,60000,80000,20000,120000,18000,9000,62000,100000,220000,150000]),
   'Defaut Status':pd.Series(['N','N','N','N','N','N','Y','Y','Y','Y','Y'])
}


df=pd.DataFrame(d)
df
```

```python
df['Age']=(df['Age']-df['Age'].mean())/(df['Age'].max()-df['Age'].min())
df['Loan']=(df['Loan Amount']-df['Loan Amount'].mean())/(df['Loan Amount'].max()-df['Loan Amount'].min())
```

```python
df.sort_values(by='Loan',ascending=True)
```

```python
Value1=int(input("Enter K Value"))
print("Prediction:",str(df['Defaut Status'][:Value1].mode())[5])
```
