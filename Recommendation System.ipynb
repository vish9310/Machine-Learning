```python
import pandas as pd
import numpy as np
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise import Reader
from surprise.model_selection import train_test_split
```

```python
reader = Reader(rating_scale=(1,10))
data=Dataset.load_from_df(rating[['User ID','Movie ID','Rating']],reader)
```

```python
[trainset, testset] = train_test_split(data,test_size=.15,shuffle=True)
recom = KNNWithMeans(k=50,sim_options={'name':'cosine','user_based':True})
recom.fit(data.build_full_trainset())     #using full training set and not doing 
test_pred=recom.test(testset)
RMSE=accuracy.rmse(test_pred)
print('RMSE  =' , RMSE )
```

```python
'''
predict () function is used nto compute the rating prediction for given user and item.

'''

pre=recom.predict (1,6) 
print("The rating Value =user 1  movie 6 is  ", pre[3])

pre=recom.predict (2,2) 
print("The rating Value =user 2  movie 2 is  ", pre[3])

pre=recom.predict (2,4) 
print("The rating Value =user 2  movie 4 is ", pre[3])

pre=recom.predict (3,1) 
print("The rating Value =user 3  movie 1 is ", pre[3])


pre=recom.predict (3,6) 
print("The rating Value =user 3  movie 6 is ", pre[3])


pre=recom.predict (4,5) 
print("The rating Value =user 4  movie 5 is ", pre[3])


pre=recom.predict (2,2) 
print("The rating Value =user 2  movie 2 is ", pre[3])

```

# book_recom Data Set

```python
book_recom=pd.read_csv("book_recom.csv")
book_recom.head()
```

```python
book = book_recom[['bookRating','userID','bookTitle']]
book.head()
```

```python
pivot1 = book.pivot_table(index='bookTitle',columns='userID',values='bookRating').fillna(0)
pivot1
```

```python
pivot1.shape
```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import sklearn
```

```python
book_matrix = csr_matrix(pivot1.values)
book_matrix
```

```python
model_knn = NearestNeighbors(metric='cosine' ,algorithm='brute')
model_knn.fit(book_matrix)
```

# Recommendation for reading

.Find the 5 similar(with maximum similarity-means lowest distance) books as that of the first book '1st to DIe':A novel'.
.The function kneighbours() is used to find k number of neighbours of a point.
.We need to compute 6 neighbours from bok 1 to all other books .Note that the shortest distance possible now is 0,which is book 1 itself .Hence,we need to take 6 neighbours,imstead of 5

```python
query_index1=0 #1st book
distances, indices = model_knn.kneighbors(pivot1.iloc[query_index1,:].values.reshape(1,-1),n_neighbors=6)
```

```python
#Display the distance
distances
```

```python
#Display respective book indices 
indices
```

```python
# as distance is a 2D array,we need to flatern it to make 1D array
```

```python
print('Recommendation for {0}\n'.format(pivot1.index[query_index1]))
for i in indices.flatten()[1:]:
    print(pivot1.index[i])
```
